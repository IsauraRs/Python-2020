# Números

En este cuaderno de jupyter vamos a ver:
1. Los diferentes números que maneja python
2. Operaciones sobre este tipo de dato


Python maneja el tipo de dato númerico, este tipo tiene varias clases dependiendo del número con el que queramos trabajar.

Ejemplos|Clase del número
-- | --
1, 2, -5, 1000|int
1.2, -0.5, 2e2,3E2|float
4+3j, 7-3.5j|complex

Lo primero que tenemos que notar es que todos estos tipos son clases

```python
help(int)
```

```python
help(float)
```

```python
help(complex)
```

```python
print(type(3))
print(type(-3))
print(type(3.1))
print(type(3e2))
print(type(3+3j))
```

Existen ciertas reglas a la hora de asignar nombres a variables:

1. Las variables no pueden iniciar con un número
2. No pueden existir espacios en el nombre, se una en su lugar guion bajo. ejem: la_variable 
3. No se puede utilizar ninguno de estos simbolos :'",<>/?|\()!@#$%^&*~-+
4. Se considera buena practica (PEP8) que los nombres sean lowercase
5. Evitar usar palabras reservadas in python como list y str
